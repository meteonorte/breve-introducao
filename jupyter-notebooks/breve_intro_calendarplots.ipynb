```python
import calplot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1 - Ler os dados diários do Rio Negro

```python
rionegro = pd.read_csv('rionegro_2020_2023.csv')
```

Veja o cabeçalho do `DataFrame`:

```python
rionegro.head()
```

Conversão da coluna `Data` para o formato `datetime`:

```python
rionegro['Data'] = pd.to_datetime(rionegro['Data'], infer_datetime_format=True)
```

Determinação de `Data` como `index` do `DataFrame`:

```python
rionegro = rionegro.set_index('Data', drop=True)
```

Veja o cabeçalho do novo `DataFrame`:

```python
rionegro.head()
```

# 2 - Criando *Calendar Plots*

**2.1 - *Calendar plot* das cotas diárias do Rio Negro**.

Exatamente o **mesmo gráfico** que você viu na coluna **Amazônia(s)** sobre as [Cheias do Rio Negro](https://www.meteonorte.com/post/rionegro).

```python
fig, ax = plt.subplots(figsize=(9, 9), nrows=4)

# a)
map1 = calplot.yearplot(rionegro['Cota (m)']['2020-01-01':'2020-12-31'], 
                        cmap='PuOr',
                        daylabels=['', '', '', '', '', '', ''],
                        monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                                     'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                        vmin=20, vmax=30, 
                        ax=ax[0])

# b) 
calplot.yearplot(rionegro['Cota (m)']['2021-01-01':'2021-12-31'], 
                 cmap='PuOr',
                 daylabels=['', '', '', '', '', '', ''],
                 monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                 vmin=20, vmax=30, 
                 ax=ax[1])

# c)
calplot.yearplot(rionegro['Cota (m)']['2022-01-01':'2022-12-31'], 
                 cmap='PuOr',
                 daylabels=['', '', '', '', '', '', ''],
                 monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                 vmin=20, vmax=30, 
                 ax=ax[2])

# d)
calplot.yearplot(rionegro['Cota (m)']['2023-01-01':'2023-04-20'], 
                 cmap='PuOr',
                 daylabels=['', '', '', '', '', '', ''],
                 monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                 vmin=20, vmax=30, 
                 ax=ax[3])

# barra de cores
cb = fig.colorbar(map1.get_children()[1], orientation='horizontal',
                  ticks=np.arange(20, 32, 2), ax=ax[3], 
                  label='Cota (m)')
cb.ax.minorticks_off()

# textos
ax[0].text(1., 10, 'Nível do Rio Negro em Manaus (AM)', fontsize=16, color='black')
ax[0].text(1., 7.2, '2020', fontsize=12, color='black')
ax[1].text(1., 7.2, '2021', fontsize=12, color='black')
ax[2].text(1., 7.2, '2022', fontsize=12, color='black')
ax[3].text(1., 7.2, '2023', fontsize=12, color='black')
ax[3].text(0, -7.7, 'Dados: Porto de Manaus', fontsize=8, color='grey')
ax[3].text(24, -7.7, 'meteonorte.com', fontsize=8, color='grey')
ax[3].text(47.2, -7.7, 'taggo.one/willyhagi', fontsize=8, color='grey')

plt.tight_layout()
#plt.savefig('rionegro_cota.jpg', dpi=300)
plt.show()
```

**2.2 - *Calendar plot* das variações das cotas diárias do Rio Negro**.

```python
fig, ax = plt.subplots(figsize=(9, 9), nrows=4)

# a)
map1 = calplot.yearplot(rionegro['Variação (cm)']['2020-01-01':'2020-12-31'], 
                        cmap='RdBu',
                        daylabels=['', '', '', '', '', '', ''],
                        monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                                     'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                        vmin=-15, vmax=15, 
                        ax=ax[0])

# b) 
calplot.yearplot(rionegro['Variação (cm)']['2021-01-01':'2021-12-31'], 
                 cmap='RdBu',
                 daylabels=['', '', '', '', '', '', ''],
                 monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                 vmin=-15, vmax=15, 
                 ax=ax[1])

# c)
calplot.yearplot(rionegro['Variação (cm)']['2022-01-01':'2022-12-31'], 
                 cmap='RdBu',
                 daylabels=['', '', '', '', '', '', ''],
                 monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                 vmin=-15, vmax=15, 
                 ax=ax[2])

# d)
calplot.yearplot(rionegro['Variação (cm)']['2023-01-01':'2023-03-31'], 
                 cmap='RdBu',
                 daylabels=['', '', '', '', '', '', ''],
                 monthlabels=['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
                              'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                 vmin=-15, vmax=15, 
                 ax=ax[3])

# barra de cores
cb = fig.colorbar(map1.get_children()[1], orientation='horizontal',
                  ticks=np.arange(-15, 17, 2), ax=ax[3], 
                  label='Variação (cm)')
cb.ax.minorticks_off()

# textos
ax[0].text(1., 10, 'Variação do Nível do Rio Negro em Manaus (AM)', fontsize=16, color='black')
ax[0].text(1., 7.2, '2020', fontsize=12, color='black')
ax[1].text(1., 7.2, '2021', fontsize=12, color='black')
ax[2].text(1., 7.2, '2022', fontsize=12, color='black')
ax[3].text(1., 7.2, '2023', fontsize=12, color='black')
ax[3].text(0, -7.7, 'Dados: Porto de Manaus', fontsize=8, color='grey')
ax[3].text(24, -7.7, 'meteonorte.com', fontsize=8, color='grey')
ax[3].text(47.2, -7.7, 'taggo.one/willyhagi', fontsize=8, color='grey')

plt.tight_layout()
#plt.savefig('rionegro_variacao.jpg', dpi=300)
plt.show()
```
